# Tight-binding energy dispersion of 1D chain
입자의 위치 집합, 에너지가 다를 때

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = 1 # hopping amplitude
a = 1 # lattice constant
c = 2 # constant

ka_list = list(np.arange(-10, 10, 0.2))
kb_list = list(np.arange(-10.1, 10.1, 0.2))

e_ka = []
e_kb = []

for k in ka_list :
    e_ka.append(-2 * t * math.cos(k * a) + c)
    
for k in kb_list:
    e_kb.append(-2 * t * math.cos(k * a) - c) # 뭔가 예상했던 그래프와 다르다...
    
plt.plot(ka_list, e_ka, label = 'e_ka')
plt.plot(kb_list, e_kb, label = 'e_kb')

plt.xlabel('k')
plt.ylabel('e_k')
plt.legend(loc = 'best')
plt.show()
```
